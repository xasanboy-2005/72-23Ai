```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree

X1 = np.array([[70], [80], [85], [90], [95], [100]])
y1 = np.array([15000, 20000, 25000, 45000, 50000, 60000])

model1 = DecisionTreeRegressor(max_depth=2)
model1.fit(X1, y1)

plt.figure(figsize=(10, 6))
plot_tree(model1, filled=True, feature_names=["House Size"], rounded=True)
plt.title("Decision Tree 1 (Depth = 2)")
plt.show()
```

```python
X2 = np.array([[60], [65], [75], [85], [95], [105]])
y2 = np.array([12000, 14000, 18000, 26000, 40000, 55000])

model2 = DecisionTreeRegressor(max_depth=3)
model2.fit(X2, y2)

plt.figure(figsize=(10, 6))
plot_tree(model2, filled=True, feature_names=["House Size"], rounded=True)
plt.title("Decision Tree 2 (New Data)")
plt.show()
```

```python
X3 = np.array([[70], [72], [75], [78], [80], [85], [90], [95], [100]])
y3 = np.array([14000, 15000, 16000, 17000, 18000, 30000, 40000, 50000, 60000])

model3 = DecisionTreeRegressor(max_depth=4)
model3.fit(X3, y3)

plt.figure(figsize=(12, 6))
plot_tree(model3, filled=True, feature_names=["House Size"], rounded=True)
plt.title("Decision Tree 3 (Depth = 4, More Data)")
plt.show()
```
