```python
# 📦 Kutubxonalarni chaqiramiz
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# 📊 Sun’iy ma’lumotlar (3ta xususiyat bilan)
X = np.array([
    [1.0, 2.1, 3.2],
    [1.5, 2.4, 2.8],
    [2.0, 1.8, 3.5],
    [2.5, 3.0, 2.2],
    [3.0, 3.2, 2.0],
    [3.5, 3.5, 1.8],
    [4.0, 2.9, 1.2],
    [4.5, 3.8, 0.9]
])

# 🏷️ Maqsadli klass (label)
y = [0, 0, 0, 0, 1, 1, 1, 1]

# 🔄 Trening/testga bo‘lamiz
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 🧠 SVM modelni yaratamiz va o‘rgatamiz
model = SVC(kernel='linear')
model.fit(X_train, y_train)

# 📈 Bashorat va tahlil
y_pred = model.predict(X_test)
print("📋 Klassifikatsiya hisobot:\n", classification_report(y_test, y_pred))

# 🎥 3D Vizualizatsiya
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Trening ma'lumotlar
for i in range(len(X_train)):
    ax.scatter(X_train[i][0], X_train[i][1], X_train[i][2], c='blue' if y_train[i]==0 else 'red', marker='o', s=60)

# Test natijalari (bashoratlar)
for i in range(len(X_test)):
    ax.scatter(X_test[i][0], X_test[i][1], X_test[i][2],
               c='cyan' if y_pred[i]==0 else 'orange', marker='^', s=80, edgecolors='k')

ax.set_xlabel("X1 o‘lcham")
ax.set_ylabel("X2 o‘lcham")
ax.set_zlabel("X3 o‘lcham")
ax.set_title("SVM modelda 3D klassifikatsiya")
plt.show()
```
